```python
population_vital = pd.merge(
    population,
    mortality,
    on=["County Code", "Year", "County"],
    how="right",
)
population_vital
```

```python
shipment.rename(columns={"BUYER_STATE": "State Abbreviation"}, inplace=True)
shipment.rename(columns={"BUYER_COUNTY": "County"}, inplace=True)
shipment.rename(columns={"YEAR": "Year"}, inplace=True)
shipment.head(5)
```

```python
population_vital.rename(columns={"State_x": "State"}, inplace=True)
population_vital.rename(columns={"State_y": "State Abbreviation"}, inplace=True)
population_vital
```

```python
merged_all = pd.merge(
    population_vital, shipment, on=["State Abbreviation", "County", "Year"], how="left"
)
merged_all
```

```python
compare_states = merged_all[
    merged_all["State"].isin(["Wisconsin", "New Jersey", "Oregon", "Washington"])
]
compare_states
```

```python
# step 1: grouping
compare_states["Group"] = compare_states["State"].apply(
    lambda x: "exp_group" if x == "Washington" else "con_group"
)

# step 2: filtering the year
filtered_data = compare_states[merged_all["Year"].between(2008, 2016)]

# step 3: see the grouped data
exp_group_data = filtered_data[filtered_data["Group"] == "exp_group"]
con_group_data = filtered_data[filtered_data["Group"] == "con_group"]
```

```python
exp_states_grouped = (
    exp_group_data.groupby("Year")
    .agg({"Deaths": "sum", "Population": "sum", "MME": "sum"})
    .reset_index()
)
con_states_grouped = (
    con_group_data.groupby("Year")
    .agg({"Deaths": "sum", "Population": "sum", "MME": "sum"})
    .reset_index()
)
```

```python
combined_data = pd.concat([exp_states_grouped, con_states_grouped])
combined_data
```

```python
combined_data.reset_index(drop=True, inplace=True)

# 'if_treat' column create
combined_data["if_treat"] = 0

# first 7 rows assigned 0 (index 0~6)
combined_data.loc[0:7, "if_treat"] = 1

# next 7 rows assigned 1 (index 7~13)
combined_data.loc[8:13, "if_treat"] = 0
combined_data
```

```python
combined_data["if_effect"] = combined_data["Year"].apply(
    lambda x: 1 if 2012 <= x <= 2015 else 0
)
combined_data
```

```python
combined_data["Opioid_Prescription_Per_Capita"] = (
    combined_data["MME"] / combined_data["Population"]
)
combined_data["Mortality_Rate_Per_Capita"] = (
    combined_data["Deaths"] / combined_data["Population"]
)
combined_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plot = sns.lmplot(
    data=combined_data[combined_data["if_treat"] == 1],
    x="Year",
    y="Opioid_Prescription_Per_Capita",
    hue="if_effect",
    legend=False,
)
plot.set(
    title=f"Pre and Post Implemenation Opioid Prescription Per Capita for Washington"
)

plt.legend(
    loc="lower center",
    bbox_to_anchor=(0.5, -0.3),
    title="Policy Implementation",
    ncol=2,
)

ax = plot.axes[0, 0]
policy_year = 2012
ax.axvline(policy_year, ls="--", color="red")
ax.set_ylabel("Opioid Prescriptions Per Capita")
```
